```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

fs = 1000
path = 'SavedData/'
#all data from https://github.com/Vyachez
```

```python
#Fast Fourier transform
def fft(data, fs):
    L = len(data)
    freq = np.linspace(0.0, 1.0 / (2.0 * fs **-1), L // 2)
    yi = np.fft.fft(data)[1:]
    y = yi[range(int(L / 2))]
    return freq, abs(y)
```

```python
# Notch filter 60Hz BP 7-13
n_60_bp_7_13 = pd.read_csv(path+'OpenBCI-RAW-test_BP_7-13.txt', sep=",", header=None).as_matrix()
channels = []
for i in range(8):
    channels.append(n_60_bp_7_13[:,1+i].astype(np.float))

for i in range(len(channels)):
    freq, y = fft(channels[i], fs)
    plt.plot(freq, y)
plt.ylim(0, 1e7)
plt.xlim(0,100)
```
